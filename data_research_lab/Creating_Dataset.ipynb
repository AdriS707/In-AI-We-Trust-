```python
# Reload the CSV file to correctly interpret the structure
file_path = 'data_trust-in-ai_2025-01-02_10-50.csv'
import pandas as pd

df = pd.read_csv(file_path, encoding='utf-16', delimiter=';')

# Drop columns which are not needed
df = df[df["MODE"] != "admin"]
df = df.drop(columns=["QUESTNNR", "MODE", "B003_01", "B009_01"])

# Correct Austrian residencies
count_removed = df[df["F005"].astype(str) == "2"].shape[0]

df["F005"].fillna(1, inplace=True)
df = df[df["F005"].astype(str) != "2"]
print(f"Anzahl der entfernten Zeilen mit Wert '2': {count_removed}")

# Drop columns where all rows after the second row are NaN
filtered_df = df.loc[:, ~(df.iloc[2:].isna().all())]

# Save the filtered DataFrame to a new CSV or overwrite the original file
output_path = "filtered_csv_file_2025-01-02_10-50_2.csv"  # Replace with your desired output file path
filtered_df.to_csv(output_path, index=False)

# Display the filtered DataFrame to verify
filtered_df


```

```python
# Reload the CSV file to correctly interpret the structure
file_path = 'data_trust-in-ai_2025-01-02_10-50.csv'
import pandas as pd

df = pd.read_csv(file_path, encoding='utf-16', delimiter=';')

# Drop columns which are not needed
df = df[df["MODE"] != "admin"]
df = df.drop(columns=["QUESTNNR", "MODE", "B003_01", "B009_01"])
df = df.drop(columns=["STARTED","TIME001","TIME002","TIME003","TIME004","TIME005","TIME006","TIME007","TIME008","TIME_SUM","LASTDATA","FINISHED","Q_VIEWER","LASTPAGE","MAXPAGE","MISSING","MISSREL","TIME_RSI", "G001_01"])

# Correct Austrian residencies
df["F005"].fillna(1, inplace=True)
df = df[df["F005"].astype(str) != "2"]


# Drop columns where all rows after the second row are NaN
analysis_df = df.loc[:, ~(df.iloc[2:].isna().all())]

# Save the filtered DataFrame to a new CSV or overwrite the original file
output_path = "analysis_2025-01-02_10-50.csv"  # Replace with your desired output file path
analysis_df.to_csv(output_path, index=False)

# Display the filtered DataFrame to verify
analysis_df
```
